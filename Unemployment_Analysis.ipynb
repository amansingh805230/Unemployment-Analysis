**Data Analysis Using Bar Chart**

```python
#Employed_in_different_regions_of_india
figure=px.bar(df,x="Region",y="Employed",color="Region",title="Employment(Region Wise)",labels={"Employed":"Employemnt Rate"})
figure.show()
```

```python
#Unemployment_in_different_regions_of_india
figure=px.bar(df,x="Region",y="Unemployment Rate",color="Region",title="Unemployment(Region Wise)")
figure.show()
```

```python
#Unemployment_in_different_states_of_india
figure=px.bar(df,x="States",y="Unemployment Rate",color="States",title="Unemployment(State Wise)")
figure.show()
```

```python
#Employment_in_different_states_of_india
figure=px.bar(df,x="States",y="Employed",color="States",title="Employment(State Wise)",labels={"Employed":"Employment Rate"})
figure.show()
```

**Data Analysis using Scatter Plot**

```python
#Unemployment_in_different_states_of_india
figure=px.scatter(df,x="States",y="Unemployment Rate",color="States",title="Unemployment(State Wise)")
figure.show()
```

```python
#Employment_in_different_states_of_india
figure=px.scatter(df,x="States",y="Employed",color="States",title="Employment(State Wise)",labels={"Employed":"Employment Rate"})
figure.show()
```

**Data Analysis using Sun Burst**

```python
#unemployment_analysis
unemp=df[["States","Region","Unemployment Rate"]]
figure=px.sunburst(unemp,path=["Region","States"],values="Unemployment Rate",title="Unemployment Rate in India")
figure.show()
```

```python
#Employment_analysis
unemp=df[["States","Region","Employed"]]
figure=px.sunburst(unemp,path=["Region","States"],values="Employed",title="Employment Rate in India")
figure.show()
```
